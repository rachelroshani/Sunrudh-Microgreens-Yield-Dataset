```python
print("Growth days out of range:", df[(df['Growth_Duration_Days'] < 5) | (df['Growth_Duration_Days'] > 25)].shape[0])
print("Temperature out of range:", df[(df['Temperature_C'] < 15) | (df['Temperature_C'] > 35)].shape[0])
print("Invalid yields:", df[df['Harvest_Yield_grams'] <= 0].shape[0])
```

```python
from scipy.stats import skew, kurtosis

numeric_cols = df.select_dtypes(include="number").columns

for col in numeric_cols:
    print(f"\n{col}")
    print("Skewness:", skew(df[col].dropna()))
    print("Kurtosis:", kurtosis(df[col].dropna()))
```

```python
numeric_cols = ["Growth_Duration_Days", "Harvest_Yield_grams", "Temperature_C", "Humidity_%"]

for col in numeric_cols:
    sns.histplot(df[col], kde=True, bins=20)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```
